```python
from math import *
import numpy as np
import matplotlib . pyplot as plt
```

```python
#valores de x
x = np.array ([1.0,1.3 ,1.6 ,1.9 ,2.2])
#valores de f(x)
f = np.array ([0.7651977, 0.6200860, 0.4554022, 0.2818186, 0.1103623])
#grado del polinomio
n = x.size
#Diferncias Cero
F= np.zeros (( n , n ) )
F[:,0]= f
#Punto a interpolar
xi =1.5
y0 = F[0 , 0]
```

```python
# Genera la tabla de diferencias
print("Diferencias 0\t1° Dif.     2° Dif.      3° Dif.      4° Dif.\t")
for i in range (0, n ):
  for j in range (1,  i+1) :
     F[i,j]=(F[i , j-1] - F[i-1, j-1]) / ( x[i] - x[ i-j ])
print(F)
```

```python
# calcula la interpolacion de xi
x0=1
for k in range (1,n) :
  x0= x0*( xi - x[k-1])
  y0= y0 + F[k,k] * x0

print ("El valor Interpolado de",xi,"es: ", y0 )
```

```python
#construccion del polinomio para graficar
x0=1
xi=np.linspace(0, 2, 70)
for k in range (1,n) :
  x0= x0*( xi - x[k-1])
  y0= y0 + F[k,k] * x0
plt.plot(xi,y0,color="red")
plt.legend(loc = 'upper right')
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.axvline(0.517757, color="grey", linestyle='dotted')
plt.axhline(0.517757, color="grey", linestyle='dotted')
plt.xlim(0, 2)
plt.ylim(0, 3)
plt.title(r'$P_4(x)$')
plt.grid()
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
#plt.savefig("output.png") # Guardar gráfico como imágen PNG.
# Mostrarlo.
plt.show()
```
